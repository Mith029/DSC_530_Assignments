```python
pmf = thinkstats2.Pmf(resp.numkdhh, label= "numkdhh")
thinkplot.Pmf(pmf)
thinkplot.Config(xlabel="Children under 18", ylabel="Pmf")
```

```python
def biasPmf(pmf,label):
    updated_pmf = pmf.Copy(label = label)
    for i,j in pmf.Items():
        updated_pmf.Mult(i,i)
        
    updated_pmf.Normalize()
    return updated_pmf
bias = biasPmf(pmf,"biased")
thinkplot.PrePlot(2)
thinkplot.Pmfs([pmf, bias])
thinkplot.Config(xlabel="Children count", ylabel="PMF")
print(f"pmf mean: {pmf.Mean()}\nbiased mean: {bias.Mean()}\n")
```

```python
pmf = pmf = thinkstats2.Pmf(resp.numkdhh, label= "numkdhh")

def PmfMean(pmf):
    mean = 0
    for i,j in pmf.Items():
        mean += (i*j)
    return mean

def PmfVar(pmf,mean):
    var = 0
    for i,j in pmf.Items():
        var += (j*(i - mean)**2)
    return var
mean = pmf.Mean()
var = pmf.Var()
computed_mean = PmfMean(pmf)
computed_var = PmfVar(pmf,mean)
print(f"Mean: {computed_mean}\nVariance: {computed_var}\n")
print("Verifying")
print(computed_mean == mean)
print(computed_var == var)
```

```python
def PercentileRank(scores, your_score):
    count = 0
    for score in scores:
        if score <= your_score:
            count += 1

    percentile_rank = 100.0 * count / len(scores)
    return percentile_rank

live, firsts, others = first.MakeFrames()

first_wgt = firsts.totalwgt_lb
first_wgt = first_wgt.dropna()
 
other_wgt = others.totalwgt_lb
other_wgt = other_wgt.dropna()

first_pmf = thinkstats2.Pmf(first_wgt, label='first')
other_pmf = thinkstats2.Pmf(other_wgt, label='other')

first_cdf = thinkstats2.Cdf(firsts.totalwgt_lb, label='first')
other_cdf = thinkstats2.Cdf(others.totalwgt_lb, label='other')

print(f"Since I was a second child, my percentile rank in 'other' category is {other_cdf.PercentileRank(7.9)}")
```
